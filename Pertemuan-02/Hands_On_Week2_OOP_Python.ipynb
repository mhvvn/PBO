## **List**

Membuat List

```python
huruf = ["a", "b", "c", "d", "e", "f"]

print(huruf[1:4])
```

```python
numbers = [1, 2, 3, 4, 5]

mixed = [10, "Python", 3.14, True]

print(numbers)
print(mixed)
```

Mnegakses Elemen List

```python
fruits = ["Apple", "Banana", "Cherry"]

print(fruits[0])
print(fruits[1])
print(fruits[-1])
```

Mengubah Elemen List

```python
fruits[1] = "Mango"
print(fruits)
```

```python
fruits.append("Orange")
print(fruits)

fruits.insert(1, "Grapes")
print(fruits)
```

Menghapus Elemen List

```python
fruits.remove("Mango")
print(fruits)

fruits.pop(2)
print(fruits)

del fruits[1]
print(fruits)
```

```python
for fruit in fruits:
    print(fruit)
```

Kesimpulan:

1. List menyimpan banyak nilai dalam satu variabel
2. Bisa diakses, diubah, ditambah, dan dihapus
3. Loop dan List Comprehension membuatnya lebih fleksibel

# **Konversi Tipe Data**

Konversi Otomatis (Implicit Conversion)

```python
angka_int = 10
angka_float = angka_int + 3.5

print(angka_float)
print(type(angka_float))
```

Python mengubah 10 (int) menjadi 10.0 (float) agar sesuai dengan operasi float.

Konversi Manual (Explicit Conversion)

int ke float / float ke int

```python
x = 5
x_float = float(x)

y = 7.8
y_int = int(y)

print(x_float)
print(y_int)
```

int/float ke string

```python
angka = 42
angka_str = str(angka)

teks = "100"
teks_int = int(teks)

print(angka_str, type(angka_str))
print(teks_int, type(teks_int))
```

list ke tuple

```python
data_list = [1, 2, 3]
data_tuple = tuple(data_list)
data_tuple2 = (4, 5, 6)
data_list2 = list(data_tuple2)

print(data_tuple, type(data_tuple))
print(data_list2, type(data_list2))
```

list/tuple ke set

```python
angka_list = [1, 2, 2, 3]
angka_set = set(angka_list)

angka_tuple = tuple(angka_set)

print(angka_set)
print(angka_tuple)
```

Dictionary ke Lis

```python
data_dict = {"nama": "Riccy", "umur": 17}

keys_list = list(data_dict.keys())

values_list = list(data_dict.values())

print(keys_list)
print(values_list)
```

Kesimpulan:

1. Konversi Otomatis terjadi saat Python perlu menyamakan tipe data dalam operasi.
2. Konversi Manual (Casting) dilakukan dengan fungsi bawaan (int(), float(), str(), list(), dll.).
3. Gunakan konversi yang tepat sesuai kebutuhan, terutama untuk menghindari error saat operasi antar tipe data.
