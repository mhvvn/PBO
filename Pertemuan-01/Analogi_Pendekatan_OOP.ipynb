# OOP

Object Oriented Programming atau yang biasa disingkat dengan OOP adalah suatu paradigma pemograman yang berfokus pada objek. Dalam pemograman OOP, kita memandang suatu program berisikan objek-objek yang saling mendukung dan berinteraksi satu sama lain. Tidak semua bahasa pemograman dapat mendukung paradigma pemograman ini dan Python adalah salah satu bahasa pemograman yang mendukung OOP.

dengan OOP kita bisa menambahkan fitur/atribut tanpa harus menulis ulang kembali programnya

```python
# Kelas dasar untuk Mie Instan
class MieInstan:
    def __init__(self, rasa):
        self.rasa = rasa

    def masak(self):
        print(f"\nMemasak mie instan rasa {self.rasa}")
        print("1. Rebus air")
        print("2. Masukkan mie")
        print("3. Tunggu 3 menit")
        self.tambahan()
        print("4. Tambahkan bumbu")
        print("5. Aduk dan sajikan!\n")

    def tambahan(self):
        pass

# Kelas untuk Mie Goreng
class MieGoreng(MieInstan):
    def tambahan(self):
        print(">> Tiriskan mie sebelum mencampur bumbu!")

# Kelas untuk Mie Kuah dengan Telur
class MieKuahTelur(MieInstan):
    def tambahan(self):
        print(">> Masukkan telur saat merebus mie!")

# Membuat objek dari kelas MieGoreng dan MieKuahTelur
mie_goreng = MieGoreng("Mie Goreng Spesial")
mie_kuah_telur = MieKuahTelur("Mie Kuah Ayam + Telur")

# Memasak masing-masing varian mie
mie_goreng.masak()
mie_kuah_telur.masak()
```

#### menambahkan atribut baru yaitu "level kepedasan" pada mie

```python
# Kelas dasar untuk Mie Instan
class MieInstan:
    def __init__(self, rasa, level_pedas="Normal"):
        self.rasa = rasa
        self.level_pedas = level_pedas  # Tambahan level kepedasan

    def masak(self):
        print(f"\nMemasak mie instan rasa {self.rasa} (Level Pedas: {self.level_pedas})")
        print("1. Rebus air")
        print("2. Masukkan mie")
        print("3. Tunggu 3 menit")
        self.tambahan()
        print("4. Tambahkan bumbu")
        print("5. Aduk dan sajikan!\n")

    def tambahan(self):
        pass

# Kelas untuk Mie Goreng
class MieGoreng(MieInstan):
    def tambahan(self):
        print(">> Tiriskan mie sebelum mencampur bumbu!")

# Kelas untuk Mie Kuah dengan Telur
class MieKuahTelur(MieInstan):
    def tambahan(self):
        print(">> Masukkan telur saat merebus mie!")

# Membuat objek dari kelas MieGoreng dan MieKuahTelur
mie_goreng = MieGoreng("Mie Goreng Spesial", "Tidak Pedas")
mie_kuah_telur = MieKuahTelur("Mie Kuah Ayam + Telur", "Sedang")

# Memasak masing-masing varian mie
mie_goreng.masak()
mie_kuah_telur.masak()

# Membuat mie instan biasa dengan level kepedasan tinggi
mie_pedas = MieInstan("Ayam Geprek", "Super Pedas")
mie_pedas.masak()
```
